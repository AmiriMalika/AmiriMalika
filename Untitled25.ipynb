**Holiday Analysis**
In this section, we will analyze the weekdays on which holidays occur each year. This helps us understand how many pre-holiday days exist in each week marked as "True" in the "IsHoliday" field.

If a particular week has more pre-holiday days in one year compared to another, sales in that week are likely to be higher. Since our model might not account for this automatically, we may need to adjust the predicted sales at the end.

Additionally, if there are very few or no regular workdays before a holiday, sales during the holiday week might be lower than the previous week.

To analyze this, we can use SQL to determine the week number of each holiday for every year. According to research:

The Super Bowl, Labor Day, and Thanksgiving always fall on the same weekday each year.
Christmas, however, is always on December 25th, meaning the weekday can vary, which may impact sales patterns.

```python
pysqldf("""
SELECT
    T.*,
    case
        when ROW_NUMBER() OVER(partition by Year order by week) = 1 then 'Super Bowl'
        when ROW_NUMBER() OVER(partition by Year order by week) = 2 then 'Labor Day'
        when ROW_NUMBER() OVER(partition by Year order by week) = 3 then 'Thanksgiving'
        when ROW_NUMBER() OVER(partition by Year order by week) = 4 then 'Christmas'
    end as Holyday,
    case
        when ROW_NUMBER() OVER(partition by Year order by week) = 1 then 'Sunday'
        when ROW_NUMBER() OVER(partition by Year order by week) = 2 then 'Monday'
        when ROW_NUMBER() OVER(partition by Year order by week) = 3 then 'Thursday'
        when ROW_NUMBER() OVER(partition by Year order by week) = 4 and Year = 2010 then 'Saturday'
        when ROW_NUMBER() OVER(partition by Year order by week) = 4 and Year = 2011 then 'Sunday'
        when ROW_NUMBER() OVER(partition by Year order by week) = 4 and Year = 2012 then 'Tuesday'
    end as Day
    from(
        SELECT DISTINCT
            Year,
            Week,
            case
                when Date <= '2012-11-01' then 'Train Data' else 'Test Data'
            end as Data_type
        FROM feat_store
        WHERE IsHoliday = True) as T""")
```

We can use Weighted Mean Absolute Error (WMAE) to evaluate performance. This is like Mean Absolute Error (MAE), but with different weights:


*   Holiday weeks are more important (weight =
5)
*   Other weeks are less important (weight = 1)


Since we can't use GridSearchCV or RandomSearchCV (which are common ways to tune models), we need to write a custom function to calculate WMAE and use it for model evaluation.

```python
def WMAE(dataset, real, predicted):
    # Assign weights: 5 for holiday weeks, 1 for regular weeks
    weights = dataset.IsHoliday.apply(lambda x: 5 if x else 1)

    # Calculate Weighted Mean Absolute Error (WMAE)
    # 1. Multiply the absolute errors by their corresponding weights
    # 2. Sum the weighted absolute errors
    # 3. Divide by the total sum of weights to normalize
    # 4. Round the result to 2 decimal places for better readability
    return np.round(np.sum(weights * abs(real - predicted)) / np.sum(weights), 2)
```

**Choosing Random Forest Regression for This Project**
For this project, we have chosen Random Forest Regression because it is a powerful ensemble method that provides high accuracy, handles missing values well, and reduces overfitting compared to individual decision trees. It works by combining multiple decision trees to make predictions, taking the average of all tree outputs for a more stable and reliable result.

**How Random Forest Works**


*   The model builds multiple decision trees
(controlled by the n_estimators parameter).
Each tree selects a feature as the root node and splits the dataset based on that feature.
*   We can choose a function to measure the quality of the split and set it as a parameter.

*   
Splitting continues until the number of samples in a node is smaller than min_samples_split, at which point it turns into a leaf node.
*   The min_samples_leaf parameter sets the minimum number of samples required to form a leaf node.

*   The max_features parameter controls the maximum number of features considered for splitting, ensuring diversity among the trees.
*   The depth of the trees is determined by the max_depth parameter, preventing excessive growth that could lead to overfitting.





**Why Use Random Forest?**


*   It reduces overfitting by averaging multiple trees, leading to better generalization.
*   It is robust to noise and missing values, making it more reliable for real-world data.

*   It handles complex relationships well, making it suitable for predicting sales trends.
*   It provides feature importance scores, helping us understand which factors influence predictions the most.


**Next Steps: Implementing Random Forest**

To ensure the best performance, we will fine-tune these parameters using cross-validation and hyperparameter tuning. The functions for implementing Random Forest, tuning parameters, and performing cross-validation will be discussed next.

```python
def random_forest(n_estimators, max_depth):
    result = []  # List to store results for different hyperparameter combinations

    # Loop through different values of n_estimators (number of trees in the forest)
    for estimator in n_estimators:
        # Loop through different values of max_depth (maximum depth of each tree)
        for depth in max_depth:
            wmaes_cv = []  # List to store WMAE values for cross-validation

            # Perform 4-fold cross-validation
            for i in range(1, 5):
                print('k:', i, ', n_estimators:', estimator, ', max_depth:', depth)

                # Split training data into train and test sets
                x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.3)

                # Create a Random Forest Regressor with given parameters
                RF = RandomForestRegressor(n_estimators=estimator, max_depth=depth)

                # Train the model on the training set
                RF.fit(x_train, y_train)

                # Make predictions on the test set
                predicted = RF.predict(x_test)

                # Calculate WMAE for the predictions and store it
                wmaes_cv.append(WMAE(x_test, y_test, predicted))

            # Print the average WMAE for the current combination of parameters
            print('WMAE:', np.mean(wmaes_cv))

            # Store the results in the list
            result.append({'Max_Depth': depth, 'Estimators': estimator, 'WMAE': np.mean(wmaes_cv)})

    # Convert results into a DataFrame for easy analysis
    return pd.DataFrame(result)
```

```python
def random_forest_II(n_estimators, max_depth, max_features):
    result = []  # List to store results of different max_features values

    for feature in max_features:
        wmaes_cv = []  # List to store WMAE scores for cross-validation

        for i in range(1,5):  # Perform 4-fold cross-validation
            print('k:', i, ', max_features:', feature)

            # Split the training data into training and testing sets
            x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.3)

            # Initialize the Random Forest Regressor with given parameters
            RF = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth, max_features=feature)

            # Train the model
            RF.fit(x_train, y_train)

            # Predict on the test set
            predicted = RF.predict(x_test)

            # Calculate and store WMAE score
            wmaes_cv.append(WMAE(x_test, y_test, predicted))

        # Print and store the average WMAE score
        print('WMAE:', np.mean(wmaes_cv))
        result.append({'Max_Feature': feature, 'WMAE': np.mean(wmaes_cv)})

    return pd.DataFrame(result)  # Return results as a DataFrame
```

```python
def random_forest_III(n_estimators, max_depth, max_features, min_samples_split, min_samples_leaf):
    result = []  # Store results for different parameter combinations

    for split in min_samples_split:
        for leaf in min_samples_leaf:
            wmaes_cv = []  # Store WMAE scores for cross-validation

            for i in range(1,5):  # Perform 4-fold cross-validation
                print('k:', i, ', min_samples_split:', split, ', min_samples_leaf:', leaf)

                # Split data into training and testing sets
                x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.3)

                # Initialize and train the Random Forest Regressor
                RF = RandomForestRegressor(n_estimators=n_estimators, max_depth=max_depth,
                                           max_features=max_features, min_samples_leaf=leaf,
                                           min_samples_split=split)
                RF.fit(x_train, y_train)

                # Predict and calculate WMAE score
                predicted = RF.predict(x_test)
                wmaes_cv.append(WMAE(x_test, y_test, predicted))

            # Store the average WMAE score for the current parameter combination
            print('WMAE:', np.mean(wmaes_cv))
            result.append({'Min_Samples_Leaf': leaf, 'Min_Samples_Split': split, 'WMAE': np.mean(wmaes_cv)})

    return pd.DataFrame(result)  # Return results as a DataFrame
```

**Adjusting for Christmas Sales Impact**
To account for the impact of Christmas on sales, adjustments are made based on historical holiday patterns. In 2010, Christmas week had 0 holiday days, while in 2011 and 2012, it had 1 and 3 holiday days, respectively. This results in a difference of 3 days from 2012 to 2010 and 2 days from 2012 to 2011, averaging 2.5 days over a full week (7 days).

To correct for this, sales from week 51 are adjusted and added to week 52 to compensate for the missing holiday effect that the model may not have captured. However, this adjustment is only applied to stores and departments where the sales difference between week 51 and week 52 is significant. Specifically, adjustments are made where sales in week 51 are more than twice the sales of week 52 (week 51 > 2 × week 52).

To efficiently implement this correction, a separate dataframe is created, and SQL is used to process the adjustments.

```python
# Create a new dataframe 'Final' containing only relevant columns from X_test
Final = X_test[['Store', 'Dept', 'Week']]

# Add a new column 'Weekly_Sales' to store the predicted sales values
Final['Weekly_Sales'] = predict
```

```python
Final_adj = pysqldf("""
    SELECT
        Store,  -- Store ID
        Dept,   -- Department ID
        Week,   -- Week number
        Weekly_Sales,  -- Original predicted weekly sales

        -- Adjust the sales for Week 52 if the previous week's sales (Week 51)
        -- were significantly higher (more than twice the sales of Week 52)
        case
            when Week = 52 and last_sales > 2 * Weekly_Sales
            then Weekly_Sales + (2.5 / 7) * last_sales  -- Apply adjustment to account for Christmas week effect
            else Weekly_Sales  -- Otherwise, keep the original Weekly_Sales
        end as Weekly_Sales_Adjusted

    from (
        -- Subquery to compute last week's sales (Week 51) for the same Store and Department
        SELECT
            Store,
            Dept,
            Week,
            Weekly_Sales,

            -- Get the sales from the previous week (Week 51) using the LAG function
            -- This will be used to compare against Week 52 sales
            case
                when Week = 52 then lag(Weekly_Sales) over (partition by Store, Dept)
            end as last_sales

        from Final  -- Using the original predicted sales data
    )
""")
```
